### Analisis de los resultados y Conclusiones

Luego de haber trabajado este proyecto para predecir el incumplimiento de pago en los productos BNPL mediante modelos de machine learning Se logró un análisis muy interesante. En primer lugar, comenzamos a identificar las características de los clientes con un alto riesgo de que probablemente no harían el pago con el fin de mejorar la toma de decisiones crediticias para las empresas tratando de minimizar las pérdidas económicas por incumplimiento de pagos y también evitando rechazar los clientes confiables. Una observación de nuestra base de datos es que normalmente estas modalidades se inclinan a prestar o dar créditos a personas con perfiles riesgosos, algo quizás peligroso para las empresas, pero tomándolo a la vida real. Se podría inducir que se les está prestando a estas personas con un alto interés que compense el riesgo que está tomando la financiera.

En primer lugar, se trabajó en las etapas de Cris-DM desde:


* Comprender al negocio

* Entender los datos

* Preparar los datos

* Modelado y entrenamiento

* Análisis o evaluación


En cada una de las fases se llevó un gran análisis Un resumen de lo que obtuvimos en cada fase es:

1) Aquí logramos entender en qué consistía el negocio, quiénes eran los involucrados. ¿A quién está afectando o qué riesgos se están tomando?

2) Realizamos el primer vistazo a la base de datos, identificando qué variables teníamos, cómo estas se comportaban con nuestra variable objetivo (análisis univariado y multivariado), además de saber cuál es su correlación frente a ella.

3) Comenzamos a hacer las transformaciones de los datos, creamos categorías para algunas variables numéricas, utilizamos dummies, definimos los pipelines, etc.

4) Esta fue la parte más fuerte del proyecto, ya que decidimos trabajar con 4 modelos de predicción que fueron: Logistic Regression, Random Forest, XGBoost y XGBoost-Ultra. Que permitieran predecir si una persona iba a incumplir o no. Midiendo métricas y comparando los modelos entre sí.

Como resultado de este proyecto, de los 4 modelos que se presentaron en este proyecto, el que más se destacó fue Random Forest con mayor Recall (0.88), logrando identificar correctamente al 88% de los clientes que incumplieron. Esto lo convierte en el más efectivo para nuestra prioridad principal: identificar riesgos altos. El siguiente a este fue XGBoost-Ultra, que obtuvo el segundo mejor Recall (0.725 - 0.773 según la versión de features) con mejor precisión y un ROC-AUC competitivo (0.775), ofreciendo un mejor balance general.


Aunque Random Forest destaca en primer lugar, hay que tener mucho cuidado en su precisión, ya que está en alrededor de 0.50. Permitiendo que el modelo genere muchos más falsos positivos, es decir, rechaza a clientes con un buen perfil. Indicando que quizás podría ser mejor tomar el segundo mejor modelo o es necesario mejorar las métricas para tener una mejor precisión.


Por ultimo para obtener mejores resultado se decidio jugar con el umbral y ver los residuales que habian donde se puede destacar que el mejor threshold es de  aproximadamente 0.2. Este maximiza el Recall, pero aumenta significativamente los rechazos de clientes confiables. El umbral de 0.27 ofrece el mejor balance entre detectar incumplidores y controlar el costo de falsos positivos.
